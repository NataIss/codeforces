```python
dfAll=pd.read_csv('ratingHistory4reds_17Feb2021.csv')
dfAll
```

```python
sMin=dfAll.groupby('handle')['newRating'].min()
```

```python
sMax=dfAll.groupby('handle')['newRating'].max()
```

```python
dfMin=sMin.to_frame(name="min")
dfMax=sMax.to_frame(name="max")
df1=pd.merge(dfMin,dfMax,left_index=True, right_index=True)
```

```python
tour=dfAll.groupby('handle')['contestId'].count()
```

```python
df1=pd.merge(df1,tour,left_index=True, right_index=True)
df1
```

```python
df2 = df1[(df1['min']<1500)]
df2
```
